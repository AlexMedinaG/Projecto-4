```python
#import dependencies 
import os
import pandas as pd
```

```python
#combining csv files
folder_path = './data/raw data'
csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]
df_list = []  # Use a different name for the list

for csv_file in csv_files:
    csv_file_path = os.path.join(folder_path, csv_file)
    temp_df = pd.read_csv(csv_file_path)
    temp_df['decade'] = csv_file
    df_list.append(temp_df)

combined_df = pd.concat(df_list, ignore_index=True) 
combined_df.head()
```

```python
#rename columns
combined_df.rename(columns={'target':'hit'}, inplace=True)
#add url for html usage
combined_df.rename(columns={'uri':'url'},inplace=True)
combined_df['url'] = combined_df['url'].apply(lambda x: f"https://open.spotify.com/track/{x.split(':')[-1]}")
#add column with decade
combined_df['decade'] = combined_df['decade'].str.extract(r'(\d+)')
combined_df
```

```python
#change decade to integer
combined_df['decade'] = combined_df['decade'].astype(int)
combined_df
```

```python
#revised data types
combined_df.dtypes
```

```python
combined_df['duration_min'] = round(combined_df['duration_ms']/60000, 3)
combined_df
```

```python
combined_df.to_csv('./data/combined_df.csv', header=True, index=False)
```

```python
#import mongo for database
from pymongo import MongoClient

df = pd.read_csv('./data/combined_df.csv')
```

```python
#sending data to Mongo database
mongo= MongoClient(port=27017)
db = mongo.SpotifySongs

SpotifySongs = df
SpotifySongs = SpotifySongs.to_dict(orient='records')

db.SpotifySongs.insert_many(SpotifySongs)
```

```python
#Import libraries for data visualization

import matplotlib.pyplot as plt
import seaborn as sns

# Sklearn import

from sklearn.model_selection import train_test_split
from sklearn import preprocessing

#Model
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVC, SVC
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

data = pd.read_csv("./data/combined_df.csv")
```

```python
# Preprocessing
# Drop Categorical Data

data.drop(['track', 'artist','url'],axis=1,inplace=True)
```

```python
#Train - Test Split
y = data['hit']
X = data.drop('hit', axis = 1)
```

```python
#Split train-test data into 7:3 (for RFC)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, shuffle = True, random_state = 1)


#Scale 

X_train = pd.DataFrame(preprocessing.robust_scale(X_train), index=X_train.index)
X_test = pd.DataFrame(preprocessing.robust_scale(X_test), index=X_test.index)
```

```python
models = {
    "                   Logistic Regression": LogisticRegression(),
    "                   K-Nearest Neighbors": KNeighborsClassifier(),
    "                         Decision Tree": DecisionTreeClassifier(),
    "Support Vector Machine (Linear Kernel)": LinearSVC(),
    "   Support Vector Machine (RBF Kernel)": SVC(),
    "                        Neural Network": MLPClassifier(),
    "                         Random Forest": RandomForestClassifier(),
    "                     Gradient Boosting": GradientBoostingClassifier()
}

for name, model in models.items():
    model.fit(X_train, y_train)
    print(name + " trained.")
```

```python
for name, model in models.items():
    print(name + ": {:.2f}%".format(model.score(X_test, y_test) * 100))
```
