```python
#import dependencies 
import os
import pandas as pd
```

```python
#combining csv files
folder_path = './data/raw data'
csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]
df_list = []  # Use a different name for the list

for csv_file in csv_files:
    csv_file_path = os.path.join(folder_path, csv_file)
    temp_df = pd.read_csv(csv_file_path)
    temp_df['decade'] = csv_file
    df_list.append(temp_df)

combined_df = pd.concat(df_list, ignore_index=True) 
combined_df.head()
```

```python
#rename columns
combined_df.rename(columns={'target':'hit'}, inplace=True)
#add url for html usage
combined_df.rename(columns={'uri':'url'},inplace=True)
combined_df['url'] = combined_df['url'].apply(lambda x: f"https://open.spotify.com/track/{x.split(':')[-1]}")
#add column with decade
combined_df['decade'] = combined_df['decade'].str.extract(r'(\d+)')
combined_df
```

```python
#change decade to integer
combined_df['decade'] = combined_df['decade'].astype(int)
combined_df
```

```python
#import mongo for database
from pymongo import MongoClient

df = pd.read_csv('./data/combined_df.csv')
```

```python
#sending data to Mongo database
mongo= MongoClient(port=27017)
db = mongo.SpotifySongs

SpotifySongs = df
SpotifySongs = SpotifySongs.to_dict(orient='records')

db.SpotifySongs.insert_many(SpotifySongs)
```
